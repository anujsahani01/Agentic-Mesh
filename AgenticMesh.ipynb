````python
"""
So we will be creating a wrapper class -- where we will be using an external LLM and make it behave like a llama_index llm.

In this case we are using 'microsoft/Phi-3-mini-4k-instruct' which will be capped with Virtual Large Language Model -- for efficient memory utilization
performance.
We will use VLLM for generating outputs. But we will override these methods with llama_index llm class.

So out llm will be able to perform all the functionalities of llama_index llm but under the hood VLLM will be making the predictions.
""";
````

````python
from llama_index.llms.vllm import Vllm  # Use this LLM class from llama_index

def initialize_engine(model_name: str):
    """Initialize the LLM"""
    from llama_index.llms.vllm import Vllm

    model = Vllm(
        model= model_name,
        tensor_parallel_size=1,                                                         # this represents the number of GPUs you have
        max_new_tokens=150,
        dtype = "float16",
        temperature = 0.1,          # earlier it was 0.9                                # good range for creative replies
        # top_p = 0.5,                # earlier it was 0.9 as well
        top_k = 50,
        vllm_kwargs={"swap_space": 1,                                                   # this allows an effcient memory managment, by allowing swapping between CPU and GPU as and when required
                     "gpu_memory_utilization": 0.7,
                     "max_model_len": 4000,
                     'enable_prefix_caching' : False,
                     'disable_sliding_window' : False,
                     'gpu_memory_utilization' :  0.7,                                   # setting GPU itilization as colab free tier has limited VRAM
                     'enforce_eager' : False,
                     'kv_cache_dtype' : 'fp8_e4m3',                                     # this will reduce the memmory utilization (scaling factors are also applied as we are changing the dtype to FP8)
                     'quantization_param_path' : "kv_scaling_factors.json",
                     'num_lookahead_slots' : 2,                                         # look ahead slots will generate answers based on future tokem so that the model just have to select which prediction will be suited best (reduce the inference time)
                     'hf_overrides' : {"use_cache":True,"attention_dropout":0.1},       # reduce repetitive compuatation / dropout will stabalize generated output
                     'tokenizer_mode' : 'auto'}
    )

    return model
````

````python
import re
from typing import Tuple

from llama_index.core.agent.react.types import (
    ActionReasoningStep,
    BaseReasoningStep,
    ResponseReasoningStep,
)
from llama_index.core.output_parsers.utils import extract_json_str
from llama_index.core.types import BaseOutputParser
````

````python
"""ReAct output parser."""

import re
import numpy as np
from typing import Tuple
from llama_index.core.agent.react.types import (
    ActionReasoningStep,
    BaseReasoningStep,
    ResponseReasoningStep,
)
from llama_index.core.output_parsers.utils import extract_json_str
from llama_index.core.types import BaseOutputParser

def extract_tool_use(input_text: str) -> Tuple[str, str, str]:
    pattern = r"\s*Thought: (.*?)\n+Action: ([^\n]+).*?\n+Action Input:(.*)"
    match = re.search(pattern, input_text, re.DOTALL)
    if not match:
        raise ValueError(f"Could not extract tool use from input text: {input_text}")

    thought = match.group(1).strip()
    action = match.group(2).strip()
    action_input = match.group(3).strip()
    return thought, action, action_input

def action_input_parser(json_str: str) -> dict:
    processed_string = re.sub(r"(?<!\w)\'|\'(?!\w)", '"', json_str)
    pattern = r'"(\w+)":\s*"([^"]*)"'
    matches = re.findall(pattern, processed_string)
    return dict(matches)

def extract_final_response(input_text: str) -> Tuple[str, str]:
    pattern = r"\s*Thought:(.*?)Answer:(.*?)(?:$)"
    match = re.search(pattern, input_text, re.DOTALL)
    if not match:
        raise ValueError(
            f"Could not extract final answer from input text: {input_text}"
        )

    thought = match.group(1).strip()
    answer = match.group(2).strip()
    return thought, answer

class CustomReActOutputParser(BaseOutputParser):
    """ReAct Output parser."""

    def semantic_similarity(self, input1, input2):
        from sentence_transformers import SentenceTransformer
        model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")
        embeddings = model.encode([input1, input2])
        similarity = np.dot(embeddings[0], embeddings[1]) / (
            np.linalg.norm(embeddings[0]) * np.linalg.norm(embeddings[1])
        )
        return similarity

    def is_action_input_matching_query(self, action_input: dict, input_text: str) -> bool:
        """Check if the action input exactly matches the user query."""
        if "input" not in action_input:
            print("Action Input Error: 'input' key missing")
            return False
        similarity_score = self.semantic_similarity(action_input.get('input'), input_text)
        return similarity_score > 0.7

    def parse_action_reasoning_step(self, output: str, input_text: str) -> BaseReasoningStep:
        """
        Parse an action reasoning step from the LLM output.
        """
        import dirtyjson as json

        thought, action, action_input = extract_tool_use(output)
        json_str = extract_json_str(action_input)

        try:
            action_input_dict = json.loads(json_str)
        except Exception as e:
            print("JSON Parsing Error:", e)
            print("Malformed JSON String:", json_str)
            raise e

        # Check if action input matches the query
        if self.is_action_input_matching_query(action_input_dict, input_text):
            # Directly return observation if there is a similarity match
            return ResponseReasoningStep(
                thought="(Implicit) Action input matches query. Returning observation...",
                response=action_input_dict.get('input', ""),
                is_streaming=False,
            )

        return ActionReasoningStep(
            thought=thought, action=action, action_input=action_input_dict
        )

    def parse(self, output: str, input_text: str, is_streaming: bool = False) -> BaseReasoningStep:
        """Parse output from ReAct agent.

        We expect the output to be in one of the following formats:
        1. If the agent needs to use a tool to answer the question:
            ```
            Thought: <thought>
            Action: <action>
            Action Input: <action_input>
            ```
        2. If the agent can answer the question without any tools:
            ```
            Thought: <thought>
            Answer: <answer>
            ```
        """

        if "Thought:" not in output:
            return ResponseReasoningStep(
                thought="(Implicit) I can answer without any more tools!",
                response=output,
                is_streaming=is_streaming,
            )

        if "Action:" in output:
            return self.parse_action_reasoning_step(output, input_text)

        if "Answer:" in output:
            thought, answer = extract_final_response(output)
            return ResponseReasoningStep(
                thought=thought, response=answer, is_streaming=is_streaming
            )

        raise ValueError(f"Could not parse output: {output}")

    def format(self, output: str) -> str:
        """Format a query with structured output formatting instructions."""
        raise NotImplementedError
````

````python
class QueryManager:
  # initialze the active query as None
    def __init__(self):
        self.active_query = None  # Tracks the currently active query

  # resets the state of active query again to None (used if model shows any divergence)
    def reset_state(self):
        self.active_query = None

  # check if the active query is related to original query if not resets active query to None
    def handle_input(self, user_input):
        if self.active_query and not self.is_related(user_input):
            self.reset_state()

        # Set the current input as the active query
        self.active_query = user_input
        return self.active_query

    def is_related(self, new_input):
        # Check if the new input is related to the active query
        if not self.active_query:
            return False
        # return T/F based on similarity
        return self.semantic_similarity(new_input, self.active_query) > 0.7

    def semantic_similarity(self, input1, input2):
        # Example semantic similarity check using embeddings
        from sentence_transformers import SentenceTransformer
        model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")
        embeddings = model.encode([input1, input2])
        similarity = np.dot(embeddings[0], embeddings[1]) / (
            np.linalg.norm(embeddings[0]) * np.linalg.norm(embeddings[1])
        )
        return similarity
````

````python
def handle_input(self, user_input):
    if self.active_query and not self.is_related(user_input):
        confirmation = input(
            "You seem to be asking a new question. Do you want to continue with this query? (yes/no): "
        )
        if confirmation.lower() != "yes":
            self.reset_state()

    # Set the current input as the active query
    self.active_query = user_input
    return self.active_query
````
